```python
print(df[df["Idade"]>=18])
```

```python
# https://pandas.pydata.org/docs/user_guide/indexing.html#boolean-indexing

print(df["Idade"]>=18)
```

```python
# https://pandas.pydata.org/docs/user_guide/indexing.html#the-where-method-and-masking

print(df.where(df["Idade"]>=18))
```

```python
# operadores lógicos https://pandas.pydata.org/docs/user_guide/indexing.html#boolean-indexing
# string https://pandas.pydata.org/docs/user_guide/basics.html#vectorized-string-methods

print(df[ (df["Idade"]>=18) & (df["Nome"].str[0]=='J') ]  )
```

```python
# https://pandas.pydata.org/docs/user_guide/groupby.html

print(df.groupby("Nome").groups)

print(df.groupby("Nome").sum())
```

```python
# https://pandas.pydata.org/docs/user_guide/dsintro.html#column-selection-addition-deletion

df["AnoNascimento"] = 2023 - df["Idade"]
print(df)
```
